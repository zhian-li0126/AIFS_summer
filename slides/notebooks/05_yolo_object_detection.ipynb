# Computer Vision Notebook 5: Object Detection with YOLO

**Goal:** use a pretrained YOLO model to detect objects in an image.

Classification answers: *What is in the image?*  
Object detection answers: *What objects are in the image, and where are they?*

YOLO stands for **You Only Look Once**. It is popular because it can detect many objects quickly.

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np
import cv2

# A small pretrained model. If the docs recommend a newer model name, replace this line.
model = YOLO('yolo11n.pt')
```

## 1. Run detection on a sample image

This downloads a small image from the Ultralytics examples. You can replace the URL with your own image path.

```python
results = model('https://ultralytics.com/images/bus.jpg')
annotated = results[0].plot()  # BGR image with boxes
annotated_rgb = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 7))
plt.imshow(annotated_rgb)
plt.axis('off')
plt.title('YOLO detections');
```

## 2. Read the detection table

Each detection has a class label, confidence score, and bounding box coordinates.

```python
for box in results[0].boxes:
    cls_id = int(box.cls[0])
    name = model.names[cls_id]
    conf = float(box.conf[0])
    xyxy = box.xyxy[0].tolist()
    print(f'{name:12s} confidence={conf:.2f} box={np.round(xyxy, 1)}')
```
