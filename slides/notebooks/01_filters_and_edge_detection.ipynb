```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve, gaussian_filter
from skimage import data, color, filters, feature, img_as_float

plt.rcParams['figure.figsize'] = (6, 4)
```

## 1. Load an image

For the lecture, we use built-in sample images so nobody needs to download anything. Later you can replace this with a greenhouse, medical, robot, satellite, or sports image.

```python
image_rgb = data.astronaut()               # RGB image: height x width x 3
image_gray = color.rgb2gray(image_rgb)     # grayscale image: height x width

print('RGB image shape:', image_rgb.shape)
print('Gray image shape:', image_gray.shape)
print('Pixel range:', image_gray.min(), 'to', image_gray.max())

plt.imshow(image_rgb)
plt.title('Original RGB image')
plt.axis('off');
```

## 2. Images are matrices

A grayscale image is a 2D table of numbers. Dark pixels have small values; bright pixels have large values.

```python
patch = image_gray[120:130, 180:190]
print(np.round(patch, 2))

plt.imshow(patch, cmap='gray', interpolation='nearest')
plt.title('A 10 x 10 pixel patch')
plt.colorbar(label='brightness')
plt.axis('off');
```

## 3. Convolution filters

A filter looks at a small neighborhood around each pixel and calculates a new value. This is one of the basic building blocks of computer vision and CNNs.

```python
box_blur = np.ones((5, 5)) / 25
sharpen = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])

blurred = convolve(image_gray, box_blur)
sharpened = np.clip(convolve(image_gray, sharpen), 0, 1)

fig, axes = plt.subplots(1, 3, figsize=(13, 4))
for ax, im, title in zip(axes, [image_gray, blurred, sharpened], ['Original', 'Box blur', 'Sharpen']):
    ax.imshow(im, cmap='gray')
    ax.set_title(title)
    ax.axis('off')
plt.tight_layout()
```

## 4. Gaussian blur

Gaussian blur uses nearby pixels with smooth weights. It reduces noise before edge detection.

```python
for sigma in [0, 1, 3, 6]:
    if sigma == 0:
        out = image_gray
    else:
        out = gaussian_filter(image_gray, sigma=sigma)
    plt.figure(figsize=(4, 3))
    plt.imshow(out, cmap='gray')
    plt.title(f'Gaussian blur sigma = {sigma}')
    plt.axis('off')
    plt.show()
```

## 5. Sobel edge detection

Sobel filters estimate how quickly brightness changes in the horizontal and vertical directions. Strong changes often correspond to object boundaries.

```python
sobel_edges = filters.sobel(image_gray)
sobel_h = filters.sobel_h(image_gray)
sobel_v = filters.sobel_v(image_gray)

fig, axes = plt.subplots(1, 4, figsize=(15, 4))
for ax, im, title in zip(axes, [image_gray, sobel_h, sobel_v, sobel_edges],
                         ['Original', 'Horizontal change', 'Vertical change', 'Sobel magnitude']):
    ax.imshow(im, cmap='gray')
    ax.set_title(title)
    ax.axis('off')
plt.tight_layout()
```

## 6. Canny edge detection

Canny edge detection uses several steps: blur the image, estimate gradients, thin the edges, and keep only edges that pass threshold rules.

```python
for sigma in [1, 2, 4]:
    edges = feature.canny(image_gray, sigma=sigma)
    plt.figure(figsize=(5, 4))
    plt.imshow(edges, cmap='gray')
    plt.title(f'Canny edges, sigma={sigma}')
    plt.axis('off')
    plt.show()
```
